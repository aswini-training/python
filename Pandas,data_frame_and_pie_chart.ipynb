```python
#Pandas is a Python library used for working with data sets.
#It has functions for analyzing, cleaning, exploring, and manipulating data.
#The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

#Why Use Pandas?
"""
Pandas allows us to analyze big data and make conclusions based on statistical theories.

Pandas can clean messy data sets, and make them readable and relevant.

Relevant data is very important in data science.
"""

#What Can Pandas Do?
"""
Pandas gives you answers about the data. Like:
Is there a correlation between two or more columns?
What is average value?
Max value?
Min value?
Pandas are also able to delete rows that are not relevant, or contains wrong values, like empty or NULL values. This is called cleaning the data.
"""
#Series: A Pandas Series is like a column in a table.
#It is a one-dimensional array holding data of any type. a = [1,7,3]

```

```python
import pandas as pd
a = [5,7,3]
b = pd.Series(a)
print(b)
```

```python
import pandas as pd
a = [5,7,3]
b = pd.Series(a)
print(b[1]) #Labels - Index value to return the specific value.
```

```python
#create labels
import pandas as pd
a = [5,7,3]
b = pd.Series(a, index = ["e","f","t"]) #assigning label for values.
print(b)
print(b["f"]) #getting an value for f label.
```

```python
#panda Series from a dictionary
import pandas as pd
name = {"X": "Charles","Y": "Sony","Z": "Zebra"} #assigning keys to values
c = pd.Series(name)
d = pd.Series(name, index = ["Y"])
print(c)
print(d)
```

```python
#DataFrames: Data sets in Pandas are usually multi-dimensional tables, called DataFrames.
#Series is like a column, a DataFrame is the whole table.

import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

myvar = pd.DataFrame(data)

print(myvar)
```

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df)
```

```python
#locate Row using index
#the DataFrame is like a table with rows and columns.
#Pandas use the loc attribute to return one or more specified row(s)
#When using [], the result is a Pandas DataFrame.

print(df.loc[[1,2]]) #we can locate 2 rows by referring index number.
```

```python
#Named Index and locate named index
#With the index argument, you can name your own indexes.

import pandas as pd
x = {
    "Student Name": ["Alex", "Brinda", "Navven"],
    "Marks": [80, 45, 78]
}
df = pd.DataFrame(x, index = [1, 2, 3])
print(df)
print(df.loc[1])
```

```python
#program to draw a pie chart using two different excel sheets.
#merge() for combining data on common columns or indices
#join() for combining data on a key column or an index
#concat() for combining DataFrames across rows or columns

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
excel1 = pd.read_excel("/content/drive/MyDrive/Bonus.xlsx")
excel2 = pd.read_excel("/content/drive/MyDrive/Student_Mark.xlsx")
df = pd.concat([excel1, excel2], ignore_index=True)
labels = df['Name']
values = df['Total']
plt.figure(figsize=(25, 10)) #width,height
plt.pie(values, labels=labels, autopct=lambda p: '{:.0f}'.format(p * sum(values) / 100),startangle = 90) #to find the percentage of each value.
#'{:.0f} - {} denotes a placeholder for a variable.
#:.0f is a format specification for a floating-point number with zero decimal places.#.0 specifies the precision, indicating the number of decimal places. In this case, it's set to zero (0), meaning no decimal places.
#ndicates that the variable being formatted is a floating-point number.
plt.legend()
plt.show()


```
